**Importing Libraries**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://books.toscrape.com/'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

**Creating empty List to store book Data**

```python
books = []
```

**Extracting Data from Website**

```python
book_list = soup.find_all('article', class_ = 'product_pod')
for book in book_list:
    title = book.h3.a['title']
    price = book.find('p', class_ = 'price_color').text.strip().replace('Â','')
    availability = book.find('p', class_ = 'instock availability').text.strip()
    rating = book.find('p', class_ = 'star-rating')['class'][1]
    link_suffix = book.h3.a['href']
    link = url + link_suffix

    books.append({
        'Title':title,
        'Price':price,
        'Availability':availability,
        'Rating':rating,
        'Link': link
    })
```

**Saving results to CSV file and Excel File**

```python
books_content = pd.DataFrame(books)
books_content.to_csv('books.csv', index = False)
books_content.to_excel('books.xlsx', index = False)
```
